2. Load the data into a DataFrame

```python
from pandas.io.json import json_normalize


# Set up an empty DataFrame 
films_df = pd.DataFrame()
page = 1

while True:
    # Make the request to the API
    res = requests.get(f'http://linserv1.cims.nyu.edu:10000/films?_page={page}')
    
    
    films = res.json()
    
    if not films:
        break
    
    # Normalize the data and append it to the DataFrame
    films_df = films_df.append(json_normalize(films))
    
    
    page += 1
    
# Print the first few rows of the DataFrame 
films_df['rt_score'] = films_df['rt_score'].astype(int)
films_df
```

```python
report = films_df.groupby('director', sort=True).agg(avg_rt_score = ('rt_score', np.mean), count = ('director', np.size)).sort_values(by='avg_rt_score', ascending=False)
report
```
