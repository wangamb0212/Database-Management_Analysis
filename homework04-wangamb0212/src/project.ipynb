* create a DataFrame


```python
import pandas as pd
df_coffee = pd.read_csv("../data/raw/Coffee_domestic_consumption.csv")
df_coffee
```

* let's see the basic information of the data

```python
df_coffee.info()
```

* transforming columns (I want to transfor the column names like 1990/91  to just 1990 for the clearity of reading)

```python
new_columns = {}
for col_name in df_coffee.columns:
    if '/' in col_name:
        new_col_name = col_name[:4]
        new_columns[col_name] = new_col_name
df_coffee = df_coffee.rename(columns=new_columns)
```

```python
df_coffee
```

* Filling in missing values(I want to fill in values if the documatation for the country has a '0', and update the 'Total_domestic_consumption' base on the updates)

```python
# if the Total_domestic_consumption is 0, let's exclude that country from consideration 
df_coffee = df_coffee[df_coffee['Total_domestic_consumption'] != 0]
```

```python
df_coffee
```

```python

df_coffee.fillna(value = 0, inplace=True)
df_coffee
```

```python
df_coffee = df_coffee.replace(0).fillna(method='ffill', axis=1).fillna(method='bfill', axis=1)
```

```python
df_coffee
```

* create a new calculated column （since we have filled the 0 in the data frame, we had to calculate a new ‘Total_domestic_consumption’ ）

```python
df_coffee['Updated_Total_domestic_consumption'] = df_coffee.iloc[:, 2:-1].sum(axis=1)
```

```python
df_coffee
```

```python
df_coffee = df_coffee.drop('Total_domestic_consumption', axis=1)
```

* Let's do a visualization (I am flitering out country that consumes Robusta coffee and make a histgram of their consumption )

```python
df_filtered = df_coffee[df_coffee['Coffee type'] == 'Robusta']
```

```python
df_filtered
```

```python
import matplotlib.pyplot as plt
```

```python

plt.bar(df_filtered['Country'],df_filtered['Updated_Total_domestic_consumption'],width = 0.75)
plt.xticks(rotation = 90)
plt.xlabel('Country Name')
plt.ylabel('Consumption on Robusta')
plt.show()
```
