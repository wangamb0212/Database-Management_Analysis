*6.shows the number of accidents that occurred at each borough

```python
# transform the 'BOROUGH' column using a function 
# any other values than 'BROOKLYN', 'QUEENS', 'BRONX', 'MANHATTAN' and 'STATEN ISLAND' will be transformed to 'OTHER'
def change_borough(borough):
    default = ['BROOKLYN', 'QUEENS', 'BRONX', 'MANHATTAN', 'STATEN ISLAND']
    if borough not in default:
        return 'OTHER'
    return borough
data['BOROUGH'] = data['BOROUGH'].map(lambda x: change_borough(x))

# count values of the transformed 'BOROUGH' column
data['BOROUGH'].value_counts()
```

```python
# visualization to allow comparison between different boroughs
data['BOROUGH'].value_counts().plot(kind='pie',autopct='%1.1f%%', legend=True)
```

* 8.what are the distributions of accidents based on the geo location (latitude & longitude)?

```python
import matplotlib.pyplot as plt

# establish a new dataframe with no (0, 0) data for location
geo_data = data[(data['LATITUDE']!=0) & (data['LATITUDE']!=0)]

# establish new dataframes to filter data for various boroughs
manhattan_data = geo_data[geo_data['BOROUGH']=='MANHATTAN']
brooklyn_data = geo_data[geo_data['BOROUGH']=='BROOKLYN']
queens_data = geo_data[geo_data['BOROUGH']=='QUEENS']
bronx_data = geo_data[geo_data['BOROUGH']=='BRONX']
staten_data = geo_data[geo_data['BOROUGH']=='STATEN ISLAND']
other_data =geo_data[geo_data['BOROUGH']=='OTHER']

# plot the data in different colors
man = plt.scatter(manhattan_data['LONGITUDE'], manhattan_data['LATITUDE'], color='red', s=3)
que = plt.scatter(queens_data['LONGITUDE'], queens_data['LATITUDE'], color='orange', s=3)
brk = plt.scatter(brooklyn_data['LONGITUDE'], brooklyn_data['LATITUDE'], color='yellow', s=3)
bro = plt.scatter(bronx_data['LONGITUDE'], bronx_data['LATITUDE'], color='green', s=3)
sta = plt.scatter(staten_data['LONGITUDE'], staten_data['LATITUDE'], color='blue', s=3)
oth = plt.scatter(other_data['LONGITUDE'], other_data['LATITUDE'], color='purple', alpha=0.05, s=3)
plt.legend((man, que, brk, bro, sta, oth), 
          ('Manhattan', 'Queen', 'Brooklyn', 'Bronx', 'Staten Island', 'Other'),
           scatterpoints=1,
           loc='upper left',
           ncol=2,
           fontsize=8)
plt.show()
```

* 10.which month did the most number of accidents occur?

```python
# create a new column called 'MONTH' to record month of the accidents
data['MONTH'] = data['CRASH DATE'].dt.month

# count values on the column 'MONTH'
month_freq = data['MONTH'].value_counts()
month_freq

```

```python
# create a new Series with month abbreviations instead of numbers
import calendar
monthabbr_freq = pd.Series(data = month_freq.values, index = month_freq.index.map(lambda x: calendar.month_abbr[x]))

# draw a bar plot to represent the frequency in terns of months
monthabbr_freq.plot(kind = 'bar')
```

* According to the plot, there are less accident in April and May.
* However, the data only includes the first half of the year, we need more data to draw conclusion on the whole year.
